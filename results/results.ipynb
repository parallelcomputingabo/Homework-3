```python
import pandas as pd
```

```python
df_2 = pd.read_csv("results-assignment2.csv")
df_2.head()
```

```python
df_3 = pd.read_csv("results-assignment3.csv")
df_3.head()
```

```python
# Combine the two Dataframes based on the 'Run', 'Case', and 'Dimensions' columns
df_combined = pd.merge(df_2, df_3, on=['Run', 'Case', 'Dimensions'])
# Select the relevant columns
df_combined = df_combined[['Run', 'Case', 'Dimensions', 'Blocked CPU (s)', 'Parallel CPU (s)', 'Naive CUDA (s)', 'Tiled CUDA (s)', 'Tiled CUDA Speedup vs. Naive CUDA (x)']]
df_combined
```

```python
# Calculate Tiled CUDA Speedup vs. Parallel CPU (x)
df_combined['Tiled CUDA Speedup vs. Parallel CPU (x)'] = df_combined['Parallel CPU (s)'] / df_combined['Tiled CUDA (s)']
df_combined
```

```python
df_combined.drop(columns=['Dimensions', 'Run'], inplace=True)
```

```python
df_combined.groupby(['Case']).mean()
```
