```python
def read_matrix_dimensions(filename):
    with open(filename, 'r') as f:
        first_line = f.readline()
    dims = first_line.strip().split()
    m = int(dims[0])
    n = int(dims[1])
    return m, n

for i in range(10):
    m, n = read_matrix_dimensions(f'Homework-3/data/{i}/input0.raw')
    n2, p = read_matrix_dimensions(f'Homework-3/data/{i}/input1.raw')
    assert n == n2, f"Mismatch dimensions in test case {i}"
    print(f"Test case {i}: m={m}, n={n}, p={p}")
```
